```python
from fuzzywuzzy import fuzz
from collections import defaultdict
from tqdm import tqdm

# separate book into sentences

def tokenize_book_to_sentences(book):
    sentences = book.split(',')
    return sentences

# create a dic, sentences is key, quotation is value

def create_sentence_dict(sentences):
    sentence_dict = defaultdict(int)
    for sentence in sentences:
        sentence_dict[sentence] += 1
    return sentence_dict


def tokenize_comments(comments):
    comment_sentences = [comment.split(',') for comment in comments]
    # Expand decomposed sentences from multiple lists into a single list
    comment_sentences = [sentence for sublist in comment_sentences for sentence in sublist]
    return comment_sentences

# update sentence dic
def update_sentence_dict(sentence_dict, comments, similarity_threshold=90):
    for comment in tqdm(comments, desc="Matching comments"):
        for sentence in sentence_dict.keys():
            if fuzz.ratio(sentence, comment) > similarity_threshold:
                sentence_dict[sentence] += 1

# find the most quoted sentence
def find_most_quoted_sentence(sentence_dict):
    most_quoted_sentence = max(sentence_dict, key=sentence_dict.get)
    return most_quoted_sentence

```

```python
#applying functions
```

```python

book = text
comments = reviews['review'].tolist()
sentences = tokenize_book_to_sentences(book)
sentence_dict = create_sentence_dict(sentences)
comment_sentences = tokenize_comments(comments)
update_sentence_dict(sentence_dict, comment_sentences)
most_quoted_sentence = find_most_quoted_sentence(sentence_dict)
print(most_quoted_sentence)
```

```python
# Convert the dictionary to a list of tuples, then sort the list
sorted_sentence_dict = sorted(sentence_dict.items(), key=lambda x: x[1], reverse=True)

# print sorted dictionary
for sentence, count in sorted_sentence_dict:
    print(f"'{sentence}': {count}")
```

```python
#open txt file
with open ('/Users/huopeiyang/Desktop/Calvino, Italo - Invisible Cities (transl. William Weaver) (2012) - libgen.li.txt') as in_file:
    text=in_file.read()
import re
regex = r"\n[A-Z]+[\s&]*[A-Z\s]+\s*[0-9]\s*\n\n(.+?)\n\n\n"
content_each=re.findall(regex, text)

import pandas as pd
book_index=pd.read_csv('/Users/huopeiyang/Desktop/thesis/book_score.csv')
key=book_index['chapter'].tolist()

dic_book={}
for index,each in enumerate(content_each):
    dic_book[key[index]]=each
```

```python
for index,header in enumerate(header_each):
    dic_book[header_name[index]]=header
    
```

```python
sentence_freq = []
for sentence, freq in sorted_sentence_dict:
    if freq > 1:
        location = None
        for title, document in dic_book.items():
            if sentence.lower() in document.lower():
                location = title
                break  # Stop looking through documents once a match is found
        if location is not None:  # Only append if a match was found
            sentence_freq.append((sentence, location, freq))
```

```python
sentence_freq
```

```python
sent_fre_location = pd.DataFrame(sentence_freq, columns=['Sentence', 'Title', 'Frequency'])
```

```python
# Group by 'Title' and get the sum of 'Frequency' for each group
grouped_df = sent_fre_location.groupby('Title')['Frequency'].sum().reset_index()

```

```python
location_frequency=grouped_df.sort_values(by=['Frequency'], ascending=False)
```
